```python
def connector_approx(mesh, conn_ix):
    adj_ixs = np.where(mesh.face_adjacency == conn_ix)[0]
    angles = np.where(np.isclose( mesh.face_adjacency_angles[adj_ixs],  0.0, atol=1e-04))
    adj_with_same = adj_ixs[angles].squeeze()
    adj_pair = mesh.face_adjacency[adj_with_same]
    vertices = mesh.faces[adj_pair]
    unique_v = np.unique(np.asarray(vertices))
    center = mesh.vertices[unique_v].mean(axis=0)
    return center


def connector_face(mesh1, mesh2):
    ch1 = mesh1.convex_hull
    ch2 = mesh2.convex_hull
    loc1, ray_ix1, tri_ix1 = ch1.ray.intersects_location([ch1.centroid], [ch1.centroid - ch2.centroid])
    loc2, ray_ix2, tri_ix2 = ch2.ray.intersects_location([ch2.centroid], [ch2.centroid - ch1.centroid])
    cent1 = connector_approx(ch1, tri_ix1)
    cent2 = connector_approx(ch2, tri_ix2)
    return cent1, cent2
```

```python
from matplotlib import pyplot
from src.geom import MeshSolid
from scipy import spatial
# msolid = {str(x.id):x for x in solids}

def remove_dangles(sol):
    sol.process()
    valid_ix = np.where(sol.area_faces > 0.5)[0]
    valid_face = sol.faces[valid_ix]  
    unf = np.unique(valid_face)
    verts = sol.vertices[unf]
    ixs = np.concatenate([np.where(unf == x)[0] for x in valid_face.reshape(-1)]).reshape(-1, 3)
    ns = MeshSolid(faces=ixs, uid=sol.id, vertices=verts, validate=True)
    if len(ns.faces) > 1:
        return ns
    
def remove_dangles_f(sol):
    sol.process()
    fcts = sol.facets
    valid_ix = np.where(sol.facets_area > 2.)[0]
    valid_face = fcts[valid_ix]  
    valid_face = np.concatenate([sol.faces[x] for x in valid_face])
    unf = np.unique(valid_face)
    verts = sol.vertices[unf]
    ixs = np.concatenate([np.where(unf == x)[0] for x in valid_face.reshape(-1)]).reshape(-1, 3)
    ns = MeshSolid(faces=ixs, uid=sol.id, vertices=verts, process=True)
    if len(ns.faces) > 1:
        return ns

def prebox(solids):
    meshes = []
    for s in solids:
        
        try:

            bx = s._as_box2()

            meshes.append(bx)
        except:
            meshes.append(None)
    return meshes
```

```python


def features(scores):  
    ixs = []
    cnt = 0
    while np.any(scores) and cnt < 2:
        sc0 = np.argmin(scores[:, 1])
        sc1 = np.argmin(scores[:, 2])
        sc2 = np.argmin(scores[:, 3])

        rank = [sc0, sc1, sc2]
        # print(rank)
        ix = Counter(rank).most_common()[0][0]
        # print(ix, scores[ix][0])
        ixs.append(int(scores[ix][0]))
        scores = np.delete(scores, ix, 0)
        cnt += 1
    return ixs


def neighbors(i):
    K = 16
    N = 8
    start_ix = i * N 
    this_box = boxed_valid[i]
    ixs_this_np = np.arange(start_ix, start_ix + len(this_box.vertices))
    tss = np.concatenate(this_box.faces[this_box.facets])
    dists, ixs = tree.query(this_box.vertices, k=K)
    mask = np.asarray([[True if j not in ixs_this_np else False for j in i ] for i in ixs], dtype=bool)

    ixs_other = ixs // N
    best_count = Counter()
    best_dists = np.zeros((6, 4))
    best_index = np.zeros((6, 4))
    
    for fct_ix, facet in enumerate(this_box.facets):
        facet_vert_ix = np.unique(this_box.faces[facet])
        sm = mask[facet_vert_ix]
        vixs_other = ixs_other[facet_vert_ix]
        dist_other = dists[facet_vert_ix]
        
        best_dists[fct_ix] = np.array([dist_other[i, sm[i, :]][0] for i in range(4)])
        best_index[fct_ix] = np.array([vixs_other[i, sm[i, :]][0] for i in range(4)])

        best_count += Counter(vixs_other[sm])

    candidates = np.unique(best_index)
    
    mx = best_count.most_common()[0][-1]

    scores, keys = [], []
    xscores = np.zeros((6, min([len(candidates), len(best_count)])))
    xscores.fill(1e6)

    for ii, (k, v) in enumerate(best_count.most_common()):
        if k in candidates:
            keys.append(k)
            ds = np.where(best_index == k)
            row, col = ds
            urows = np.unique(row)
 

            cc = best_dists[ds]
            score = [k, (1. - v / mx), cc.mean(), cc.min()]
            scores.append(score)
            
    res = features(np.asarray(scores).copy())
    col = []
    for r in res:
        dists = spatial.distance.cdist(this.facets_centroids, BX[76].facets_centroids)
        ind = np.unravel_index(np.argmin(dists, axis=None), dists.shape)[0]
        col.append(ind)
    return res, col
```

```python
importlib.reload(src.geom)
importlib.reload(src.viper)
importlib.reload(src.structs.node)

from src.structs.node import Node, GeomNode, GeomEdge
from src.structs.edge import EdgeSolid
import pandas as pd
import src.propogate.solid as gs

import itertools
import trimesh

importlib.reload(gs)

def stat_edge_list(edge_list):
    ALL_NODES = []
    for eg in edge_list:
        if eg is not None:
            ALL_NODES.append(eg.source)
            ALL_NODES.append(eg.target)
    return ALL_NODES


def reduce_nodes(node_list):
    ds = []
    withgeom = []
    for n in node_list:
        ds.extend([e.curve.length for e in n.neighbors(edges=True) if e.has_geom is False])
        withgeom.extend([e.curve.length for e in n.neighbors(edges=True) if e.has_geom is True])
    return ds, withgeom

        
def sample_system(system, box_valid, ix=None, n=5):
    ix = ix if ix else np.random.choice(list(range(len(box_valid))))
    out = []
    q = [ix]
    adj_temp = defaultdict(set)
    _seen = set()
    maps = {} # {old : new}
    
    while len(out) < n and q:
        this_ix = q.pop(0)
        
        if this_ix not in maps:
            maps[this_ix] = len(out)
        
        out.append(box_valid[this_ix])
        
        _seen.add(this_ix)
        for ind in system.adj[this_ix]:
            adj_temp[this_ix].add(ind)
            if ind not in _seen:
                q.append(ind)
    new_system = src.viper.SolidSystem()
    new_system.adj  = defaultdict(set)
    for k, vs in adj_temp.items():
        new_system.facets[maps[k]] = system.facets[k]

        for v in vs:
            if v in maps:
                new_system.adj[maps[k]].add(maps[v])
    return new_system, out


def confidence_by_neighs(system, box_valid):
    """
    pt_conf = 1 - ( sum(confidence_of_neigh ** 2) - this.confidence ) 
    if two neighs are high confidence, and this is low confince, defer to neighbors
        step 1 - build node graph  
            conn_facet_centroid_point -> Node( )
            ds, dt = Node( f1 ) , Node( f1 )
            E = ds.connect_to(dt, obj=MeshSolid)

    Returns:
    ---------------
        list(Edge)
    """
    edges = [None] * len(system.adj)
    for i, nix in system.adj.items():
        this_box = box_valid[i]
        facets_ix = np.asarray(list(system.facets[i]))
        px = src.geom.primary_axis(this_box, facets_ix)
                
        if px is None:
            px = this_box.facets_centroids[facets_ix]
        n1, n2 = src.structs.Node(tuple(px[0].tolist())) , src.structs.Node(tuple(px[1].tolist()))
        E = n1.connect_to(n2, edge_cls=EdgeSolid, obj=box_valid[i])
        for ni in nix:
            if edges[ni] is not None:
                
                spnt = edges[ni].source.as_point
                epnt = edges[ni].target.as_point
                dists = [spnt.distance(n1.as_point), spnt.distance(n2.as_point),
                         epnt.distance(n1.as_point), epnt.distance(n2.as_point)
                         ]
                min_dist = min(dists)
                if min_dist == dists[0]:
                    edges[ni].source.connect_to(n1, edge_cls=EdgeSolid, obj=None)
                elif min_dist == dists[1]:
                    edges[ni].source.connect_to(n2, edge_cls=EdgeSolid, obj=None)
                elif min_dist == dists[2]:
                    edges[ni].target.connect_to(n1, edge_cls=EdgeSolid, obj=None)
                elif min_dist == dists[3]:
                    edges[ni].target.connect_to(n2, edge_cls=EdgeSolid, obj=None)
        edges[i] = E  
    return edges



def node_show_fn(node, handle='xg', **kwargs):
    if node.has_geom is True:
        src.ui.visual.to_solid(node.obj, handle=handle + '/gs', opacity=0.5)
        h_id = str(node.obj.id)
    else:
        h_id = str(node.id)
    
    for edge in node.neighbors(edges=True):
        source = np.array(list(edge.source.geom))
        target = np.array(list(edge.target.geom))
        viz3.viz_bone(source, target, handle='{}/{}/{}'.format(handle, 'skel', h_id))
    return node


def geom_show_fn(edge, handle='xg', **kwargs):
    if edge.has_geom is True:
        src.ui.visual.to_solid(edge.obj, handle=handle + '/gs', opacity=0.5)
        h_id = str(edge.obj.id)
    else:
        h_id = str(edge.id)
    source = np.array(list(edge.source.geom))
    target = np.array(list(edge.target.geom))
    viz3.viz_bone(source, target, handle='{}/{}/{}'.format(handle, 'skel', h_id))
    return edge


def nearest_facets(this, other):
    r, dists, tid = trimesh.proximity.closest_point(other, this.facets_centroids)
    armi = np.unravel_index(np.argmin(dists, axis=None), dists.shape)
    return armi


def confidence_by_nodes(system, box_valid):
    
    nodes = [None] * len(system.adj)
    def add_to_list(i):
        if nodes[i] is None:
            obj_clean = box_valid[i]
            obj_node = src.structs.node.GeomNode(obj_clean.centroid, obj=obj_clean)
            for facet_index in system.facets[i]:
                px = src.structs.node.GeomNode(
                    tuple(obj_clean.facets_centroids[facet_index].tolist()), facet_ix=facet_index 
                ) 
                obj_node.connect_to(px)
            nodes[i] = obj_node
            return obj_node
        else:
            return nodes[i]

    for i, nix in system.adj.items():
        obj_node = add_to_list(i)
        
        for ix, other_index in enumerate(system.adj[i]):
            
            other = add_to_list(other_index)
            res = nearest_facets(obj_node.obj, other.obj)
            # other.connect_to(obj_node)
            print(i, other_index, res)
            # px.connect_to(others[i+1])
            # nodes[i] = [obj_node] + others
    return list(itertools.chain(*filter(None, nodes)))
```

```python
from rtree import index


def make_index():
    p = index.Property()
    p.dimension = 3
    p.dat_extension = 'data'
    p.idx_extension = 'index'
    p.storage = index.RT_Memory
    p.overwrite = True
    idx3 = index.Index(properties=p)
    return idx3


def to_bbox(obj):
    mn = obj.vertices.min(axis=0).tolist()
    mx = obj.vertices.max(axis=0).tolist()
    return tuple(mn + mx)


def to_index(indx, nodes):
    for i, nd in enumerate(nodes):
        indx.insert(i, to_bbox(obj), obj=i)
```
