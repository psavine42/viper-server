```python
def resolve_3(this_ix, sphere_end, it, io, so):
    """ ray from this.sphere_origin, to other1.sphere_inters 
      does it intersect other2 ? 
      
    determine the case if a ray starts from this.nearest_point(that) 
     if it intersects others, before that, it should not be used 
     (that is in between this and other)
      """
    nexts = []
    this_sphere = vs2.spheres_for(this_ix)[sphere_end][:3]
    
    print(it, so, io)
    candidates =  io[np.where(it == sphere_end)]
    other_sphr =  so[np.where(it == sphere_end)]
   
    for cand_ix in candidates:
        cand_sph = other_sphr[np.argwhere(candidates != cand_ix).squeeze(axis=-1)]
        inters = subs[cand_ix].ray.intersects_location(
            np.tile(this_sphere, (cand_sph.shape[0], 1)),  
            vs2._spheres[cand_sph][:, :3]
        )
        if len(inters[0]) > 0:
            rm = candidates[np.argwhere(candidates != cand_ix).squeeze(axis=-1)][inters[1]]
            print('remove ', rm, ' from ', this_ix)
            Xconn[this_ix].add(cand_ix) 
            nexts += rm.tolist()
            
        #print('origin from {}, mesh: {}, targets: {}, has inters {}'.format( 
        #      this_ix, cand_ix, 
        #      candidates[np.argwhere(candidates != cand_ix).squeeze(axis=-1)],
        #      len(inters)))
    # print('REM', nexts)
    for n in set(nexts):
        it = it[np.argwhere(io != n).squeeze(axis=-1)]
        so = so[np.argwhere(io != n).squeeze(axis=-1)]
        io = io[np.argwhere(io != n).squeeze(axis=-1)]
    return it, so, io

def resolve_ends(this_ix):
    start, end = vs2.sphere_indexes_for(this_ix)
    # print(vs2.spheres_for(this_ix))
    print(start, end)
    it, io, so = vs2.intersections_for(this_ix)
    if start == end -1:
        it, io, so = resolve_3(this_ix, 0, it, io, so)
    else:
        it, io, so = resolve_3(this_ix, 0, it, io, so)
        it, io, so = resolve_3(this_ix, end - start -1, it, io, so)
```

```python
import src.geom as geom
import src.geombase as geomf
importlib.reload(geom)
# sample = fls[3990535]
# ds = geom.remove_dangles(sample)

def asline(smp):
    xg = []
    for e in smp.points[smp.edges_unique]:
        p1, p2 = e.tolist()
        xg.append(sg.LineString([p1, p2]))
    return sg.MultiLineString(xg)


def asboxline(smp):
    xg = []
    bbx = smp.bounding_box_oriented
    centr = bbx.centroid.tolist()
    vert_face = bbx.vertices[bbx.faces[bbx.facets]]
    fct_cnts = vert_face.reshape(bbx.facets.shape[0], -1, 3).mean(axis=1)
    for fct in fct_cnts:
        xg.append(sg.LineString([centr, fct.tolist()]))
    return sg.MultiLineString(xg)


def viz_mls(i, x, hnd):
    for j, l in enumerate(x.geoms):
        p1, p2 = l.coords
        viz3.viz_line(p1, p2, handle=viz3.as_path(hnd, i, j))
        

def as_spheres(smp):
    """
    Degenerate cases : what if  1 <  axes are same length? 
    """
    bbx = smp.bounding_box_oriented
    
    centr = bbx.centroid
    vert_face = bbx.vertices[bbx.faces[bbx.facets]]
    
    # facet centroids np.shape([6, 3])
    fct_cnts = vert_face.reshape(bbx.facets.shape[0], -1, 3).mean(axis=1)
    # fct_axes = np.asarray([[centr, fct] for fct in fct_cnts])
    
    fct_norm = bbx.facets_normal
    mn1, mn2 = np.where(spatial.distance.cdist(fct_norm, -fct_norm) == 0.)
    
    # pairs of axis points of obb : [ 3, 2]
    pairs = np.unique([sorted(v) for v in zip(mn1, mn2)], axis=0)

    # length of each axis indexed : np.shape([3]) 
    axis_lens = np.sqrt(np.sum((fct_cnts[pairs][:, 0, :] - fct_cnts[pairs][:, 1, :]) ** 2, axis=1))
    num_lens = np.unique(axis_lens)
    print(axis_lens)
    # if len(num_lens) == 3:
    axis_ords = np.argsort(axis_lens)
    middle_axis_ix = np.argwhere(axis_ords == 1).squeeze()
    longst_axis_ix = np.argwhere(axis_ords == 2).squeeze()

    num_sphere = max([1, int(axis_lens[longst_axis_ix] // axis_lens[middle_axis_ix])])
    
    if num_sphere == 1:
        sphere_rad = axis_lens[middle_axis_ix] 
        return np.expand_dims(np.concatenate([centr, [sphere_rad]], axis=0), axis=0)
    else: 
        sphere_rad = axis_lens[middle_axis_ix] 
        # print(num_sphere, sphere_rad)
        longest_axis = fct_cnts[pairs[longst_axis_ix]]
        # print(longest_axis)
        offset = (longest_axis[0, :] - longest_axis[1, :]) / num_sphere 

        print(offset)
        sphere_points =  longest_axis[1, :] + np.asarray([i*offset for i in range(num_sphere+1)])
        xt = np.tile(sphere_rad, (sphere_points.shape[0], 1)) 

        return np.concatenate([sphere_points, xt], axis=1)
    

#pts1 = as_spheres(train_ids[3497346])
#print(pts1)

#pts2 = as_spheres(train_ids[3675234]) #3828064])
#print(pts2)

pts3 = as_spheres(train_ids[3828064]) #])
print(pts3)
# geom.sphere_intersections(pts1, pts2)
```

```python
from rtree import index


def make_index():
    p = index.Property()
    p.dimension = 3
    p.dat_extension = 'data'
    p.idx_extension = 'index'
    p.storage = index.RT_Memory
    p.overwrite = True
    idx3 = index.Index(properties=p)
    return idx3


def to_bbox(obj):
    mn = obj.vertices.min(axis=0).tolist()
    mx = obj.vertices.max(axis=0).tolist()
    return tuple(mn + mx)


def to_index(indx, nodes):
    for i, nd in enumerate(nodes):
        indx.insert(i, to_bbox(obj), obj=i)
```
